## 1.  Find 10 countries with most projects

## 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# exploring 'mjtheme_namecode' column
json_df['mjtheme_namecode'].head()
```

```python
# 'mjtheme_namecode' is a json list of dict with 2 colomns:  'name' and 'code'.  We create a new DataFrame with columns 'name' and 'code': p_themes
p_themes = pd.DataFrame(columns = ['name', 'code'])

# We interrate each row in 'mjtheme_namecode' and append to p_themes
for row in json_df.mjtheme_namecode:
    p_themes = p_themes.append(pd.json_normalize(row))
p_themes.head(15)
```

```python
# names of projects can be accessed thru the 'name' column, counted and sorted in decending value, excluding where the column is blank
p_themes_names = p_themes[p_themes.name != ''].name.value_counts().sort_values(ascending = False)

# taking head(10) to return the 10 most common major projects
p_themes_names.head(10)
```

## 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Explore unique project name and code
p_themes.head(10)
```

```python
# Explore unique project name and code. 
p_themes.describe()
```

```python
# We find that there are name has 1 more unique instance, this is most likely when that name is blank.  

# Further exploring unique project name and code, we find unique instances excluding when name is blank.
p_themes[p_themes.name != ''].nunique()
```

```python
# Count instances where name is blank in the DataFrame: count_blanks
count_blanks = p_themes[p_themes.name == ''].name.value_counts()
count_blanks
```

```python
# The interesting thing is we find that there are 11 unique instances of both project name and code, yet there are 122 instances where the name is left blank.  
```

```python
# It's likely that each unique code correspond to a unique name. If that's the case, we could use drop_duplicates to create a new DataFrame of just unique corresponding code and name columns: P_Unique
P_Unique = p_themes[p_themes.name != ''].drop_duplicates()
P_Unique
```

```python
# We merge the 2 DataFrames p_themes and P_Unique into a new DataFrame to get fill blank spaces with corresponding names: P_Filled_Blanks
P_Filled_Blanks = pd.merge(p_themes, P_Unique, on = 'code', suffixes=('_y',''), how = 'right')

# Create a new DataFrame with only the name and code columns dropping the original name column which has the blanks: P_Fllled_Blanks
P_Filled_Blanks = P_Filled_Blanks[['name', 'code']]
P_Filled_Blanks
```

```python
# Explore the new DataFrame
P_Filled_Blanks.describe()
```

```python
# The new DataFrame has all the original rows with idential count of rows, instances and frequencies for both columns.  Now with count intances where the name column is still blank
count_blanks = P_Filled_Blanks[P_Filled_Blanks.name == ''].name.value_counts()
count_blanks
```
